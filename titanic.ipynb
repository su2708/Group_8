```python
import seaborn as sns
import pandas as pd

# 타이타닉 데이터셋 불러오기
titanic = sns.load_dataset('titanic')
```

```python
# 데이터셋의 처음 5개 행 확인
titanic.head()
```

```python
# 데이터셋의 기술적 통계 정보 확인
print(titanic.describe())
```

```python
# 결측치 확인
print(titanic.isnull().sum())
```

```python
titanic.describe()

# count: 각 열에 대해 유효한 값(NaN이 아닌 값)의 개수를 나타냅니다.

# mean: 각 열의 평균값을 나타냅니다.

# std: 각 열의 표준편차를 나타냅니다. 표준편차는 데이터의 분산 정도를 나타내며, 값이 클수록 데이터가 평균에서 멀리 퍼져 있음을 의미합니다.

# min: 각 열의 최소값을 나타냅니다.

# 25%: 각 열의 1사분위수(Q1)를 나타냅니다. 전체 데이터의 25%가 이 값보다 작거나 같습니다.

# 50%: 각 열의 중앙값(median)을 나타냅니다. 전체 데이터의 50%가 이 값보다 작거나 같습니다.

# 75%: 각 열의 3사분위수(Q3)를 나타냅니다. 전체 데이터의 75%가 이 값보다 작거나 같습니다.

# max: 각 열의 최대값을 나타냅니다.
```

```python
# 'age' 열의 결측치를 중앙값으로 대체
titanic['age'] = titanic['age'].fillna(titanic['age'].median())

# 'embarked' 열의 결측치를 최빈값으로 대체
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])

# 결측치 처리 확인
print("Age 열의 결측치 개수:", titanic['age'].isnull().sum())
print("Embarked 열의 결측치 개수:", titanic['embarked'].isnull().sum())
```

```python
# 수치형 인코딩
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic['alive'] = titanic['alive'].map({'no': 1, 'yes': 0})
titanic['embarked'] = titanic['embarked'].map({'C': 0, 'Q': 1, 'S': 2})

# 인코딩 결과 확인
print("sex", titanic['sex'].head())
print("alive", titanic['alive'].head())
print("embarkd", titanic['embarked'].head())
```

```python
# 새로운 feature 생성: family_size
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1

# 새로운 feature 확인
print(titanic['family_size'].head())
```
